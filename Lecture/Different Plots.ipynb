```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(0, 100, 50)
y = np.random.randint(0, 100, 50)
```

Box Plot

```python
plt.subplot(121)
plt.boxplot(x)
plt.subplot(122)
plt.boxplot(y)
plt.show()
```

Bar plot

```python
plt.bar(x,y)
plt.show()
```

Histogram Plot

```python
plt.hist(x)
plt.show()
```

Scatter Plot

```python
plt.scatter(x,y)
plt.show()
```
