```python
# Recursive fucntion for factorial

def factorial(n):
    if(n == 0 or n == 1):
        return 1
    else:
        return n * factorial(n-1)
    
print(factorial(5))
```

```python
# Fibonnaci sequence using Generator Function

def fibonacci_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b


fib_gen = fibonacci_generator(10)
for num in fib_gen:
    print(num,end=" ")
```

```python
# Plot the graphs for trigonometric functions sin, cos, tan, cot, sec & cosec for the values pi to 2pi.

import matplotlib.pyplot as plt

plt.figure(figsize=(20,16))

x = np.linspace(np.pi, 2 * np.pi, 100)
plt.subplot(2, 3, 1)
plt.plot(x, np.sin(x), label="sin(x)", color="red")
plt.title("sin(x)")
plt.xlabel("Angle")
plt.ylabel("Value")

plt.subplot(2, 3, 2)
plt.plot(x, np.cos(x), label="cos(x)", color="green")
plt.title("cos(x)")
plt.xlabel("Angle")
plt.ylabel("Value")

plt.subplot(2, 3, 3)
plt.plot(x, np.tan(x), label="tan(x)", color="blue")
plt.title("tan(x)")
plt.xlabel("Angle")
plt.ylabel("Value")

plt.subplot(2, 3, 4)
plt.plot(x, 1/np.sin(x), label="cosec(x)", color="red")
plt.title("cosec(x)")
plt.xlabel("Angle")
plt.ylabel("Value")

plt.subplot(2, 3, 5)
plt.plot(x, 1/np.cos(x), label="sec(x)", color="green")
plt.title("sec(x)")
plt.xlabel("Angle")
plt.ylabel("Value")

plt.subplot(2, 3, 6)
plt.plot(x, 1/np.tan(x), label="cot(x)", color="blue")
plt.title("cot(x)")
plt.xlabel("Angle")
plt.ylabel("Value")
```

```python
# Consider you want create dataset with ages of people in your surroundings. Use input method to ask user their age, store those ages in appropriate data type. Apply error handling that will not accept more than 130  or less than 0 inputs, raise appropriate prompts to guide users.

def collect_ages():
    ages = []
    while True:
        try:
            age = input("Enter age (or type 'done' to finish): ")
            if age.lower() == "done":
                break
            age = int(age)
            if age < 0 or age > 130:
                print("Age must be between 0 and 130. Please try again.")
                continue
            ages.append(age)
        except ValueError:
            print("Invalid input! Please enter a valid number for age.")
    return ages

age_dataset = collect_ages()
print("\nCollected Ages:", age_dataset)
```

```python
# Create two 3d arrays as matrices. Perform matrix operations (Addition, Multiplication, dot product, inverse, determinant) on those matrices. Explain identity matrix, multiply each matrix with identity matrix and record the observation. (All operations should be done with Numpy library)

import numpy as np

A = np.arange(1, 28)
A = A.reshape(3, 3, 3)
B = np.arange(29, 56)
B = B.reshape(3, 3, 3)


# Addition
C = A + B
print("Addition Result:\n", C)

# Multiplication
D = A * B
print("\nMultiplication Result:\n", D)

# Dot Product
E = np.dot(A[0, 0], B[0, 0])
print("\nDot Product Result:\n", E)

# Inverse
inv_A = [
    np.linalg.inv(A[i]) if np.linalg.det(A[i]) != 0 else "Singular"
    for i in range(A.shape[0])
]
inv_B = [
    np.linalg.inv(B[i]) if np.linalg.det(B[i]) != 0 else "Singular"
    for i in range(B.shape[0])
]
print("\nInverse Result of A:\n", inv_A)
print("Inverse Result of B:\n", inv_B)

# Determinant
det_A = [np.linalg.det(A[i]) for i in range(A.shape[0])]
det_B = [np.linalg.det(B[i]) for i in range(B.shape[0])]
print("\nDeterminant Result of A:", det_A)
print("Determinant Result of B:", det_B)

# Multiply with Identity Matrix
identity_matrix = np.eye(3)
print("\nIdentity Matrix (3x3):")
print(identity_matrix)

# Identity Matrix Multiplication
AI = np.array(
    [np.matmul(A[i], identity_matrix) for i in range(A.shape[0])]
)
BI = np.array(
    [np.matmul(B[i], identity_matrix) for i in range(B.shape[0])]
)

print("\nA * I\n", AI)
print("B * I\n", BI)

print("\nObservations:")
print(
    "Multiplying any matrix slice with the identity matrix returns the original matrix slice, as the identity matrix acts like a neutral element in matrix multiplication."
)
```
